```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gyro_output = pd.read_csv("gyro_reading_movement.csv")
```

```python
gyro_output
```

```python
gyro_output['index'] = range(len(gyro_output))
```

```python
gyro_output_cpy = gyro_output
```

```python
temperature = gyro_output.temperature
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = [20, 10]
```

```python
plt.plot(gyro_output["index"],gyro_output["ang_rate_x"],label="ang_rate_x")
plt.plot(gyro_output["index"],gyro_output["ang_rate_y"],label="ang_rate_y")
plt.plot(gyro_output["index"],gyro_output["ang_rate_z"],label="ang_rate_z")
plt.legend()
plt.show()
```

```python
plt.plot(gyro_output["index"],gyro_output["temperature"],label="temperature")
plt.plot(gyro_output["index"],[gyro_output.temperature.mean()]*len(gyro_output),label="temperature_mean")
plt.legend()
plt.show()
```

```python
plt.plot(gyro_output["index"],gyro_output.temperature.map(lambda p: p - gyro_output.temperature.mean()),label="temperature_mean_deviation")
plt.legend()
plt.show()
```
